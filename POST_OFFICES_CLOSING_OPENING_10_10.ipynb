```python
import pandas as pd
import geopandas as gpd
import datetime
```

```python
# reading the file

arquivo = "C:/Users/AnaLíviadosSantosMaz/Downloads/us-post-offices-random-coords_.csv"

us_post_offices_raw = pd.read_csv(arquivo, delimiter=";")

us_post_offices_raw
```

```python
us_post_offices_raw = us_post_offices_raw.dropna(subset=['Established'])
```

```python
# getting only the columns that are needed

us_post_offices_raw_cleaned = us_post_offices_raw.loc[:,['Name','GNIS.State', 'State', 'County1', 'GNIS.OrigCounty','Established','Discontinued','Continuous','Duration','Latitude','Longitude','RandomCoordsFlag']]

us_post_offices_raw_cleaned
```

```python
# adjusting null values by combining two complementary columns

us_post_offices_raw_cleaned['GNIS.State'] = us_post_offices_raw_cleaned['GNIS.State'].fillna(us_post_offices_raw_cleaned['State'])
us_post_offices_raw_cleaned['GNIS.OrigCounty'] = us_post_offices_raw_cleaned['GNIS.OrigCounty'].fillna(us_post_offices_raw_cleaned['County1'])

us_post_offices_raw_cleaned
```

```python
# filters the DataFrame `us_post_offices_raw_cleaned` to include only rows that meet the following criteria:
# 1. the "Established" year (founding year) is between 1600 and 2000, inclusive.
# 2. the "Discontinued" year (closing year) satisfies one of the following conditions:
#    a) it is greater than or equal to 1600.
#    b) it is null (pd.isnull checks for NaN values, indicating that the post office is still active or the year is not recorded).

us_post_offices_raw_cleaned = us_post_offices_raw_cleaned[
    ((us_post_offices_raw_cleaned['Established'] >= 1600) & (us_post_offices_raw_cleaned['Established'] <= 2000)) &
    ((us_post_offices_raw_cleaned['Discontinued'] >= 1600) | (pd.isnull(us_post_offices_raw_cleaned['Discontinued'])))
]

us_post_offices_raw_cleaned
```

```python
us_post_offices_raw_cleaned = us_post_offices_raw_cleaned.dropna(subset=['Latitude', 'Longitude'])

us_post_offices_raw_cleaned
```

```python
# convert the 'Established' and 'Discontinued' columns to numeric, handling invalid values
us_post_offices_raw_cleaned['Established'] = pd.to_numeric(us_post_offices_raw_cleaned['Established'], errors='coerce')
us_post_offices_raw_cleaned['Discontinued'] = pd.to_numeric(us_post_offices_raw_cleaned['Discontinued'], errors='coerce')

years = list(range(1790, 2001, 10))
```

```python
gdf = gpd.read_file("C:/Users/AnaLíviadosSantosMaz/Downloads/tl_2023_us_state_simplified.geojson")

gdf.head(10)
```

```python
gdf = gdf[['STUSPS', 'geometry', 'INTPTLAT','INTPTLON']]

gdf.head()
```

Difference between openings and closures

```python
# creating a DataFrame to store the results
state_yearly_counts_diff = []

# iterating through each state and year
for state in us_post_offices_raw_cleaned['GNIS.State'].unique():
    state_data = us_post_offices_raw_cleaned[us_post_offices_raw_cleaned['GNIS.State'] == state]
    previous_count = 0  # Initial count of post offices
    for year in years:
        # filter the post offices established up to the current year
        current_count = state_data[(state_data['Established'] <= year) & 
                                   ((state_data['Discontinued'] > year) | 
                                    (pd.isnull(state_data['Discontinued'])))].shape[0]
        # calculate the difference compared to the previous year
        new_openings = current_count - previous_count
        # for the year 1790, use the total up to that year
        if year == 1790:
            new_openings = current_count
        # save the results
        state_yearly_counts_diff.append({'State': state, 'Year': year, 'New_Openings': new_openings})
        previous_count = current_count

# create the final DataFrame with the results
state_yearly_counts_df_diff = pd.DataFrame(state_yearly_counts_diff)

# displaying the first few results
state_yearly_counts_df_diff
```

```python
# creating a new 'Date' column in the DataFrame by converting the 'Year' column into a datetime object.
# each year is converted to a specific date: January 1st of that year at 12:00:00 PM.
state_yearly_counts_df_diff['Date'] = state_yearly_counts_df_diff['Year'].apply(
    lambda x: datetime.datetime(x, 1, 1, 12, 0, 0)
)
```

```python
# merging the DataFrame `state_yearly_counts_df_diff` with the GeoDataFrame `gdf` to include geometric data (e.g., state boundaries).
# the merge uses the 'State' column from `state_yearly_counts_df_diff` and the 'STUSPS' column from `gdf` as the matching keys.
# a left join is performed, ensuring all rows from `state_yearly_counts_df_diff` are kept, even if no match is found in `gdf`.
df_merged_post_offices_geometry_diff = pd.merge(state_yearly_counts_df_diff, gdf, right_on='STUSPS', left_on='State', how='left')


df_merged_post_offices_geometry_diff
```

```python
df_merged_post_offices_geometry_diff['country'] = 'US'

df_merged_post_offices_geometry_diff
```

```python
df_merged_post_offices_geometry_diff = df_merged_post_offices_geometry_diff.drop(columns=['STUSPS'])

df_merged_post_offices_geometry_diff
```

```python
# save to CSV with geometry
output_path_diff = 'tests_project_us_post_offices_states/post_offices_diference_pattern_geometry.csv'
df_merged_post_offices_geometry_diff.to_csv(output_path_diff, index=False, encoding='utf-8')
```

```python
# save to CSV without geometry
df_merged_post_offices_diff = df_merged_post_offices_geometry_diff[['State','Year',	'New_Openings',	'Date',	'country']]

output_path_diff = 'tests_project_us_post_offices_states/post_offices_diference_pattern.csv'
df_merged_post_offices_diff.to_csv(output_path_diff, index=False, encoding='utf-8')
```

Only openings

```python
# creating a DataFrame to store the results
state_yearly_counts_opening = []

# iterating through each state and year
for state in us_post_offices_raw_cleaned['GNIS.State'].unique():
    # filter the data for the current state
    state_data = us_post_offices_raw_cleaned[us_post_offices_raw_cleaned['GNIS.State'] == state]
    previous_count = 0  # Initial count of post offices
    for year in years:
        # filter the post offices established up to the current year
        current_count = state_data[state_data['Established'] <= year].shape[0]
        # calculate the number of new openings compared to the previous year
        new_openings = current_count - previous_count
        # for the year 1790, use the total up to that year
        if year == 1790:
            new_openings = current_count
        # save the results
        state_yearly_counts_opening.append({'State': state, 'Year': year, 'New_Openings': new_openings})
        previous_count = current_count

# create the final DataFrame with the results
state_yearly_counts_df_opening = pd.DataFrame(state_yearly_counts_opening)

# displaying the first few results
state_yearly_counts_df_opening
```

```python
# adding a 'Date' column to the DataFrame by converting the 'Year' column into datetime objects.
# each year is represented as January 1st of that year at 12:00 PM.
state_yearly_counts_df_opening['Date'] = state_yearly_counts_df_opening['Year'].apply(
    lambda x: datetime.datetime(x, 1, 1, 12, 0, 0)
)
```

```python
# merging the DataFrame `state_yearly_counts_df_opening` with the GeoDataFrame `gdf` to include geometric data (e.g., state boundaries).
# the merge uses the 'State' column from `state_yearly_counts_df_opening` and the 'STUSPS' column from `gdf` as the matching keys.
# a left join is performed, keeping all rows from `state_yearly_counts_df_opening`, even if no match is found in `gdf`.
df_merged_post_offices_geometry_opening = pd.merge(state_yearly_counts_df_opening, gdf, right_on='STUSPS', left_on='State', how='left')

df_merged_post_offices_geometry_opening
```

```python
df_merged_post_offices_geometry_opening = df_merged_post_offices_geometry_opening.drop(columns=['STUSPS'])

df_merged_post_offices_geometry_opening
```

```python
# save to CSV
output_path_opening = 'tests_project_us_post_offices_states/post_offices_opening_pattern.csv'
df_merged_post_offices_geometry_opening.to_csv(output_path_opening, index=False, encoding='utf-8')
```

Only Closures

```python
# filtering the DataFrame to exclude rows where 'Discontinued' is zero, i.e., keeping only the closed post offices
us_post_offices_raw_cleaned = us_post_offices_raw_cleaned[us_post_offices_raw_cleaned['Discontinued'] != 0]

# creating a list to store the results
state_yearly_closures = []

# iterating through each state and year
for state in us_post_offices_raw_cleaned['GNIS.State'].unique():
    # filter the data for the current state
    state_data = us_post_offices_raw_cleaned[us_post_offices_raw_cleaned['GNIS.State'] == state]
    previous_count = 0  # Initial count of closed post offices
    for year in years:
        # filter the closed post offices up to the current year
        current_count = state_data[state_data['Discontinued'] <= year].shape[0]
        # calculate the difference in closures compared to the previous year
        new_closures = current_count - previous_count
        # for the year 1790, use the total count up to that year
        if year == 1790:
            new_closures = current_count
        # save the results
        state_yearly_closures.append({'State': state, 'Year': year, 'New_Closures': new_closures})
        previous_count = current_count

# create the final DataFrame with the results
state_yearly_df_closures = pd.DataFrame(state_yearly_closures)

# displaying the first few results
state_yearly_df_closures
```

```python
# adding a 'Date' column to the DataFrame by converting the 'Year' column into datetime objects.
# each year is represented as January 1st of that year at 12:00 PM.
state_yearly_df_closures['Date'] = state_yearly_df_closures['Year'].apply(
    lambda x: datetime.datetime(x, 1, 1, 12, 0, 0)
)
```

```python
# merging the DataFrame `state_yearly_df_closures` with the GeoDataFrame `gdf` to include geometric data (e.g., state boundaries).
# the merge uses the 'State' column from `state_yearly_df_closures` and the 'STUSPS' column from `gdf` as the matching keys.
# a left join is performed, keeping all rows from `state_yearly_df_closures`, even if no match is found in `gdf`.
df_merged_post_offices_geometry_closures = pd.merge(state_yearly_df_closures, gdf, right_on='STUSPS', left_on='State', how='left')


df_merged_post_offices_geometry_closures
```

```python
df_merged_post_offices_geometry_closures = df_merged_post_offices_geometry_closures.drop(columns=['STUSPS'])

df_merged_post_offices_geometry_closures
```

```python
# save to CSV
output_path_closures = 'tests_project_us_post_offices_states/post_offices_closures_pattern.csv'
df_merged_post_offices_geometry_closures.to_csv(output_path_closures, index=False, encoding='utf-8')
```

Combining openings and closures

```python
# merge dataframes for openings, closures, and geometry
df_combined = pd.merge(
    state_yearly_counts_df_opening,
    state_yearly_df_closures,
    on=["State", "Year", "Date"],
    how="outer"
)

# fill NaN values with 0 for calculations
df_combined['New_Openings'] = df_combined['New_Openings'].fillna(0)
df_combined['New_Closures'] = df_combined['New_Closures'].fillna(0)

# calculate net change (openings - closures)
df_combined['Net_Change'] = df_combined['New_Openings'] - df_combined['New_Closures']

# add geometry by merging with the GeoDataFrame
df_combined = pd.merge(df_combined, gdf, left_on="State", right_on="STUSPS", how="left")

# drop unnecessary columns
df_combined = df_combined.drop(columns=['STUSPS'])
```

```python
# adding a 'Change_Magnitude' column to the DataFrame that contains the absolute value of the 'Net_Change' column.
# this column represents the magnitude of the change, without considering whether it's positive or negative.
df_combined['Change_Magnitude'] = df_combined['Net_Change'].abs()

# adding a 'Change_Type' column that indicates 'Negativo' (Negative) if 'Net_Change' is less than 0, 
# otherwise it assigns 'Positivo' (Positive) if 'Net_Change' is greater than or equal to 0.
df_combined['Change_Type'] = df_combined['Net_Change'].apply(lambda x: 'Negativo' if x < 0 else 'Positivo')

# displaying the first few rows of the updated DataFrame
df_combined
```

```python
# save to CSV
output_path_combined = 'tests_project_us_post_offices_states/post_offices_combined_pattern.csv'
df_combined.to_csv(output_path_combined, index=False, encoding='utf-8')

df_combined
```
